```python
import pandas as pd
import numpy
```

```python
df = pd.read_csv('SayacOlcumleri.csv')
```

```python
df
```

```python
voltage = 220
```

```python
df['Guc_W'] = df['Olcum_Deger'] * voltage
```

```python
df
```

```python
df['Olcum_Zaman'] = pd.to_datetime(df['Olcum_Zaman'])
```

```python
df['Olcum_Saat'] = df['Olcum_Zaman'].dt.floor('H')
```

```python
hourly_sum = df.groupby(['Sayac_ID', 'Olcum_Saat'])['Guc_W'].sum().reset_index()
```

```python
hourly_sum.head()
```

```python
df.head(10)
```

```python
grouped_df = df.groupby(['Sayac_ID', 'Olcum_Saat'])['Guc_W'].sum().reset_index()
```

```python
print(grouped_df.head())
```

```python
grouped_df
```

```python
grouped_df['Day'] = grouped_df['Olcum_Saat'].dt.day
grouped_df['Month'] = grouped_df['Olcum_Saat'].dt.month
grouped_df['Year'] = grouped_df['Olcum_Saat'].dt.year
grouped_df['Hour'] = grouped_df['Olcum_Saat'].dt.hour
```

```python
grouped_df.drop(columns=['Olcum_Saat'], inplace=True)
```

```python
grouped_df.head(60)
```

```python
columns = ['Sayac_ID', 'Guc_W', 'Hour', 'Day', 'Month', 'Year']
grouped_df = grouped_df[columns]
```

```python
grouped_df
```

```python
grouped_df.to_csv('guc_w.sum.seperated.csv', index=False)
```
