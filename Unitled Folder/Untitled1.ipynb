```python
import pandas as pd
```

```python
df = pd.read_csv('ekimhavadurumu.csv')
```

```python
df
```

```python
df.columns = ['time', 'temperature(°C)', 'humidity(%)', 'cloudcover(%)', 'windspeed(km/h)','irrediance']
```

```python
df = df.iloc[2:]
```

```python
df
```

```python
df = df.drop(columns=['irrediance'])
```

```python
df
```

```python
# Convert the 'time' column to datetime format
df['time'] = pd.to_datetime(df['time'])

# Update the year to 2024
df['time'] = df['time'].apply(lambda x: x.replace(year=2024))
```

```python
df
```

```python
df.to_csv('ekimhavadurumu2024.csv', index=False)
```
