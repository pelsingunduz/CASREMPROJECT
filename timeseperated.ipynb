```python
weather_data = pd.read_csv('weather2.csv')
```

```python
print(weather_data.head())
```

```python
weather_data = weather_data.drop(columns=['Unnamed: 0'])
```

```python
weather_data['time'] = pd.to_datetime(weather_data['time'])  
```

```python
weather_data['hour'] = weather_data['time'].dt.hour
weather_data['day'] = weather_data['time'].dt.day
weather_data['month'] = weather_data['time'].dt.month
weather_data['year'] = weather_data['time'].dt.year
```

```python
weather_data = weather_data.drop(columns=['time'])
```

```python
columns = ['hour', 'day', 'month', 'year', 'temperature(°C)', 'humidity(%)', 'cloudcover(%)', 'windspeed(km/h)', 'irrediance(W/m²)']
weather_data = weather_data[columns]
```
