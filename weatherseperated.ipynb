```python
current_dir = os.path.dirname(os.path.abspath('weather.csv'))
file_path = os.path.join(current_dir,'weather.csv')
```

```python
df = pd.read_csv(file_path)
```

```python
df['time'] = pd.to_datetime(df['time']) 
```

```python
df['hour'] = df['time'].dt.hour
df['day'] = df['time'].dt.day
df['month'] = df['time'].dt.month
df['year'] = df['time'].dt.year
```

```python
df = df.drop(columns=['time'])
```

```python
df
```

```python
columns = ['hour', 'day', 'month', 'year', 'temperature(°C)', 'humidity(%)', 'cloudcover(%)', 'windspeed(km/h)', 'irrediance(W/m²)']
df = df[columns]
```

```python
output_file_path = os.path.join(current_dir, 'weather_seperated.csv')
df.to_csv(output_file_path, index = False)
```
