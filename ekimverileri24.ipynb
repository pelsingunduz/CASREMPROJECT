```python
import pandas as pd
```

```python
df = pd.read_csv('weather (1).csv')
```

```python
df
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
start_date = '2023-10-01'
end_date = '2023-10-31'
ekim_verileri = df[(df['time'] >= start_date) & (df['time'] <= end_date)]
```

```python
ekim_verileri 
```

```python
ekim_verileri['time'] = df['time'].apply(lambda x: x.replace(year=2024))
```

```python
ekim_verileri
```

```python
ekim_verileri.to_csv('ekim_verileri.csv', index=False)
```
