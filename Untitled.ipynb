```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
consumption_df = pd.read_csv('energy_consumption.date.csv')
weather_df = pd.read_csv('weather_seperated (2).csv')
```

```python
print(consumption_df.columns)
print(weather_df.columns)
```

```python
df = pd.merge(consumption_df, weather_df, 
                     on=['year', 'month', 'day', 'hour'])
```

```python
df.head()
```

```python
hourly_data= df.groupby('hour').mean()
```

```python
#Saatlik tüketetim ve sıcaklık eğilimi
plt.figure(figsize=(14,7))
sns.lineplot(x=hourly_data.index,y=hourly_data['Consumption_kWh'], 
             label='Average Consumption (kWh)')
sns.lineplot(x=hourly_data.index, y=hourly_data['temperature(°C)'],
             label='Average Temperature(°C)	', color='orange')
plt.xlabel('Hour of the Day')
plt.ylabel('Value')
plt.title('Average Hourly Electricity Consumption and Temperature')
plt.grid(True)
plt.legend()
plt.show()
```

```python
monthly_data= df.groupby('month').mean()
```

```python
#
plt.figure(figsize=(14,7))
sns.barplot(x=monthly_data.index,y=monthly_data['Consumption_kWh'], color='blue',
             label='Average Consumption (kWh)')
plt.twinx()
sns.lineplot(x=monthly_data.index, y=monthly_data['humidity(%)'], color='green',
             label='Average Humidity(%)')
plt.xlabel('Month')
plt.title('Average Monthly Electricity Consumption and Humidity')
plt.grid(True)
plt.show()
```
